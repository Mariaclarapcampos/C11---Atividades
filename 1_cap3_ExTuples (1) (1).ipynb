### ExTuples – Q1: Músicas favoritas

1. Crie uma tupla chamada `musicas_favoritas` com **5 músicas** que você mais gosta
2. Exiba a primeira e a última música usando índices
3. Exiba as 3 músicas do meio usando *slicing*
4. Mostre o tamanho da tupla com `len()`

````python
# ExTuples – Q1
# Crie sua tupla com 5 músicas favoritas
musicas_favoritas = ('Believer','Speak Now','Gateway car','August','The Archer')

# Primeira música
print(musicas_favoritas[0])

# Última música
print(musicas_favoritas[-1])

# As 3 músicas do meio (slicing)
print(musicas_favoritas[1:4])

# Tamanho da tupla
len(musicas_favoritas)
````

### ExTuples – Q2: Filmes preferidos

1. Crie uma tupla chamada `filmes` com **4 filmes** que você recomendaria a um amigo
2. Exiba os filmes em **ordem alfabética** com `sorted()`
3. Descubra em qual índice está o seu favorito com `.index()`
4. Itere com `for` imprimindo cada filme no formato: `"Recomendo: <filme>"`

````python
# ExTuples – Q2
# Crie sua tupla com 4 filmes favoritos
filmes = ('Como Treinar Seu Dragão','A Paixão de Cristo', 'Minions','Truque de Mestre')

# Filmes em ordem alfabética
ordem_alfa = sorted(filmes)
print(ordem_alfa)

# Índice do seu filme favorito
print(filmes.index('Como Treinar Seu Dragão'))

# Iteração com mensagem
for n in filmes:
  print('Iterando ' + n)
````

### ExTuples – Q3: Notas do semestre

1. Crie duas tuplas: `notas_p1` e `notas_p2`, cada uma com **3 notas** (reais ou fictícias) de matérias diferentes
2. Concatene as duas em `todas_notas` usando `+`
3. Exiba a nota máxima e mínima com `max()` e `min()`
4. Escolha uma nota e conte quantas vezes ela aparece com `.count()`

````python
# ExTuples – Q3
# Crie as tuplas com suas notas
notas_p1 = ('50','20')
notas_p2 = ('40','25')

# Concatenação
todas_notas = notas_p1 + notas_p2
print(todas_notas)

# Nota máxima e mínima
print(max(todas_notas), min(todas_notas))

# Contagem de uma nota específica
print(todas_notas.index('20'))
````
