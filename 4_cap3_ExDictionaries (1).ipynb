## ExDictionaries – Exercícios com Dicionários 📒

**Sintaxe:**
```python
meu_dict = {'chave1': 'valor1', 'chave2': 42}
```
> Dicionários são criados com `{ chave: valor }` e permitem **índices personalizados** (não numéricos).

### ExDictionaries – Q1: Perfil pessoal

1. Crie um dicionário chamado `meu_perfil` com as chaves: `nome`, `idade`, `cidade`, `curso` e `hobbie_favorito` — preenchidas com seus dados reais
2. Acesse e exiba seu `nome` e `hobbie_favorito`
3. Adicione a chave `linguagem_favorita` com o valor correspondente
4. Atualize sua `idade` para um ano a mais
5. Exiba todas as **chaves** com `.keys()` e todos os **valores** com `.values()`

````python
# ExDictionaries – Q1
# Crie seu perfil com seus dados reais
meu_perfil = {
    'nome': 'Maria Clara' ,
    'idade': 21,
    'cidade': 'Santa Rita do Sapucaí',
    'curso': 'Software',
    'hobbie_favorito': 'Desenhar' ,
}

# Acessar nome e hobbie_favorito
print(meu_perfil['nome'])
print(meu_perfil['hobbie_favorito'])

# Adicionar linguagem_favorita
meu_perfil['linguagem_favorita'] = 'JavaScript'

# Atualizar idade
meu_perfil['idade'] = 22

# Exibir todas as chaves
print(meu_perfil.keys())

# Exibir todos os valores
print(meu_perfil.values())
````

### ExDictionaries – Q2: Cardápio do restaurante favorito

1. Crie um dicionário chamado `cardapio` onde as **chaves** são pratos do seu restaurante favorito e os **valores** são os preços — inclua pelo menos **5 pratos**
2. Exiba o preço de um prato específico
3. Adicione um novo prato ao cardápio
4. Remova um prato com `del`
5. Itere com `.items()` exibindo: `"Prato: X – R$ Y"`

````python
# ExDictionaries – Q2
# Crie o cardápio do seu restaurante favorito
cardapio = {
    'prato1': 'batata frita',
    'preço1': 10,
    'prato2': 'peixe',
    'preço2': 15,
    'prato3': 'churrasco',
    'preço3': 20,
    'prato4': 'maionese',
    'preço4': 25,
    'prato5': 'salada',
    'preço5': 30,
}

# Exibir preço de um prato específico
print(cardapio['preço4'])

# Adicionar novo prato
cardapio['prato6'] = 'sopa'

# Remover um prato
del cardapio['prato3']
del cardapio['preço3']

# Iterar com .items()
for k, v in cardapio.items():
    print(f'{k} é {v}')
````

### ExDictionaries – Q3: Lista de contatos

**Sintaxe de lista de dicionários:**
```python
contatos = [
    {'nome': 'Ana', 'cidade': 'SP'},
    {'nome': 'Bruno', 'cidade': 'RJ'},
]
```

1. Crie uma lista de dicionários chamada `contatos` com **3 amigos ou colegas**, cada um com as chaves: `nome`, `telefone`, `email` e `cidade` (pode usar dados fictícios)
2. Exiba o **nome** do primeiro contato: `contatos[0]['nome']`
3. Exiba o **email** do segundo contato
4. Itere sobre a lista exibindo nome e cidade de cada contato
5. Adicione a chave `aniversario` ao terceiro contato

````python
# ExDictionaries – Q3
# Crie sua lista de contatos
contatos = [
    {
        'nome': 'Maria',
        'email': 'maria@gmail.com',
        'cidade': 'Itajubá',
        'aniversario': '08/03'
    },
    {
        'nome': 'João',
        'email': 'joao@gmail.com',
        'cidade': 'Pouso Alegre',
        'aniversario': '07/06'
    },
    {
        'nome': 'Ana',
        'email': 'ana@gmail.com',
        'cidade': 'Santa Rita do Sapucaí',

    }
]

# Nome do primeiro contato
print(contatos[0]['nome'])

# Email do segundo contato
print(contatos[1]['email'])

# Iterar exibindo nome e cidade
for contato in contatos: #é uma lista
    print(contato['nome'], '-', contato['cidade'])

# Adicionar aniversario ao terceiro contato
contatos[2]['aniversario'] = '06/06'
````
