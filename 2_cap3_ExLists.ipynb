### ExLists – Q1: Lista de desejos

1. Crie uma lista chamada `lista_desejos` com **4 itens** que você quer comprar ou conquistar (game, viagem, gadget, curso...)
2. Adicione mais **2 itens** ao final com `append()`
3. Insira um item **na posição 2** com `insert()`
4. Remova o item que você menos quer com `remove()`
5. Exiba a lista final

````python
# ExLists – Q1
# Crie sua lista de desejos com 4 itens
lista_desejos = ['game', 'viagem', 'carro', 'casa']

# Adicionar 2 itens ao final
lista_desejos.append('celular')

# Inserir item na posição 2
lista_desejos.insert(2,'gadget')

# Remover o item que você menos quer
lista_desejos.remove('carro')

# Exibir lista final
print(lista_desejos)
````

### ExLists – Q2: Placar de um jogo

1. Crie uma lista chamada `placar` com **5 entradas** no formato `'Nome: Pontuação'` (use nomes de amigos, personagens ou jogadores reais)
2. Ordene o placar em **ordem alfabética** com `sort()`
3. Reverta a ordem com `sort(reverse=True)`
4. Substitua o jogador da posição `0` por um novo usando atribuição direta
5. Exiba o placar final

````python
# ExLists – Q2
# Crie seu placar com 5 jogadores
placar = ['Maria','Lara','Letícia','Tati','Dani']

# Ordenar alfabeticamente
placar.sort()
print(placar)

# Ordenar em ordem inversa
placar.sort(reverse=True)
print(placar)

# Substituir jogador da posição 0
placar[1]='Clara'

# Exibir placar final
print(placar)
````

### ExLists – Q3: Séries para assistir

1. Crie uma lista chamada `series` com **6 séries** que você quer ou já assistiu
2. Exiba as **3 primeiras** usando *slicing*
3. Remova a série da posição `1` com `pop()` e guarde o valor em uma variável
4. Delete a última série com `del`
5. Exiba a série removida pelo `pop()` e a lista resultante

````python
# ExLists – Q3
# Crie sua lista com 6 séries
series = ['The chosen', 'Greys Anatomy', 'Lost', 'Elementary', 'O Mentalista', 'Lioness']

# As 3 primeiras séries (slicing)
print(series[0:3])

# Remover série na posição 1 e guardar o valor
removida = series.pop(0)

# Deletar a última série
del series[4]

# Exibir série removida e lista resultante
print('Série removida:', removida)
print('Lista restante:', series)
````
