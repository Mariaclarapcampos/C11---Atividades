### ExSets – Q1: Gêneros musicais em comum

1. Crie um set `meus_generos` com os gêneros musicais que **você** gosta
2. Crie um set `generos_amigo` com os gêneros que **um amigo** gosta (compartilhe alguns em comum)
3. Descubra todos os gêneros que pelo menos um de vocês gosta (**União `|`**)
4. Descubra os gêneros que **só você** gosta (**Diferença `-`**)
5. Descubra os gêneros que **ambos** gostam (**Interseção `&`**)

````python
# ExSets – Q1
# Crie seus sets com gêneros musicais
meus_generos = {'pop', 'indie', 'folkie','gospel'}
generos_amigo = {'indie', 'mpb', 'folkie', 'rock'}

# União (todos os gêneros)
todos_generos = meus_generos | generos_amigo
print(todos_generos)

# Diferença (só você gosta)
dif = meus_generos - generos_amigo
print(dif)

# Interseção (ambos gostam)
inter = meus_generos & generos_amigo
print(inter)
````

### ExSets – Q2: Limpando duplicatas do histórico

1. Crie uma **lista** chamada `historico_filmes` com pelo menos **8 filmes**, repetindo pelo menos **3 títulos** de propósito
2. Converta a lista para um **set** chamado `filmes_unicos` para eliminar as repetições
3. Exiba quantos filmes únicos você tem com `len()`
4. Verifique se um filme específico está no set usando `in`

````python
# ExSets – Q2
# Crie sua lista com filmes repetidos de propósito
historico_filmes = {'Minions' , 'Toy Story', 'Minions', 'Barbie', 'Mulan', 'Barbie', 'Rambo', 'Toy Story'}

# Converter para set
filmes_unicos = set(historico_filmes)
print(filmes_unicos)

# Quantidade de filmes únicos
print(len(filmes_unicos))

# Verificar se um filme está no set
print('Rambo' in filmes_unicos)
````

### ExSets – Q3: Habilidades da equipe

1. Crie um set `minhas_skills` com suas habilidades técnicas (linguagens, ferramentas, etc.)
2. Crie um set `skills_colega` com habilidades de um colega (compartilhe algumas)
3. Exiba habilidades que **só você tem** (Diferença)
4. Exiba habilidades que **só o colega tem** (Diferença inversa)
5. Exiba habilidades que **ambos dominam** (Interseção)
6. Exiba o conjunto completo de habilidades da equipe (União)

````python
# ExSets – Q3
# Crie os sets com suas habilidades e as do colega
minhas_skills = {'C', 'javascript', 'python', 'html', 'css'}
skills_colega = {'C', 'julia', 'java', 'R', 'json', 'html'}

# Só você tem
minhas = minhas_skills - skills_colega
print(minhas)

# Só o colega tem
colega = skills_colega - minhas_skills
print(colega)

# Ambos dominam
ambos = minhas_skills | skills_colega

# Equipe completa
print(ambos)
````
